```python
from qdrant_client import QdrantClient

# Подключение к вашему инстансу
client = QdrantClient(url="http://3.99.208.202:6333")

# Проверка подключения
print("✅ Connected!" if client.get_collections() else "❌ Connection failed")

# Список всех коллекций
collections = client.get_collections()
print("📦 Collections:", [c.name for c in collections.collections])
```

```python
# Создать снапшот
collection_name = "vacancies"  # ← замените на имя вашей коллекции
snapshot_info = client.create_snapshot(collection_name=collection_name)
print(f"📸 Snapshot created: {snapshot_info.name}")

# URL для скачивания
snapshot_url = f"http://3.99.208.202:6333/collections/{collection_name}/snapshots/{snapshot_info.name}"
print(f"🔗 Download URL: {snapshot_url}")

# Скачать файл
import requests
response = requests.get(snapshot_url)
with open(f"{collection_name}_backup.snapshot", "wb") as f:
    f.write(response.content)
print(f"✅ Downloaded: {collection_name}_backup.snapshot")
```
